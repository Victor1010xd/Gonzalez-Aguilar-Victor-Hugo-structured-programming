## 1. Core Concepts
What is a Dictionary?

A dictionary is a data structure that stores information as key-value pairs.

### General structure:
- key -> value
#### Example:
- "name" -> "Alice"
- "age" -> 20
- "major" -> "Computer Science"

Unlike lists, which store elements using numerical positions (indexes), dictionaries store and retrieve information using unique keys.

### Why Do We Need Dictionaries?
Consider the following example using lists:



````python
students = ["Alice", "Bob", "Charlie"]
grades = [95, 80, 100]
````

To find Charlie's grade:



````python
index = students.index("Charlie")
print(grades[index])
````



This approach becomes inefficient and difficult to maintain as the amount of data grows.

Using a dictionary:


````python

grades = {
    "Alice": 95,
    "Bob": 80,
    "Charlie": 100
}

print(grades["Charlie"])
````

The dictionary directly associates each student with a grade.



## Key Concepts

# Dictionary vs List

| List                 | Dictionary               |
| -------------------- | ------------------------ |
| Uses indexes         | Uses keys                |
| Ordered sequence     | Key-value mapping        |
| Access by position   | Access by identifier     |
| Example: students[0] | Example: student["name"] |

Understanding this distinction is the most important learning outcome of this lab.

````python
grades = {}
for i in range(3):
    name = input("Name: ")
    grade = int(input("Grade: "))
    grades[name] = grade
    print()

print(grades)
````

## Workspace

````python
character = {
    "name": "Alucard",
    "health": 100,
    "attack": 85,
    "defense": 70,
    "level": 1
}
print("--- Initial Statistics ---")
for key, value in character.items():
    print(f"{key.capitalize()}: {value}")
character["level"] += 1
character["attack"] += 10
print("\n--- Updated Statistics ---")
for key, value in character.items():
    print(f"{key.capitalize()}: {value}")
````

## Critical Thinking Questions
1. What is the main difference between a list and a dictionary?

Answer: **A list is an ordered sequence where elements are accessed sequentially using numerical positions (indexes starting from 0). A dictionary is an unordered collection of key-value pairs where elements are accessed directly through unique, meaningful identifiers called keys.**

2. Why are dictionaries more suitable than lists for storing student grades associated with names?

Answer: **Because dictionaries eliminate the need to maintain two separate lists or search for indexes. They provide a direct, logical connection ($O(1)$ lookup time complexity) between a student's name (the key) and their actual grade (the value), making the code cleaner and much more efficient.**

3. Give one real-world example where a dictionary would be more useful than a list.

Answer: **A user authentication system or database registry, where a unique username (the key) is mapped directly to a set of user details such as email, hashed password, and permission levels (the values).**

4. What would happen if two different values were assigned to the same key?

Answer: **Dictionary keys must be strictly unique. If you assign a new value to an existing key, the old value is overwritten (updated) with the new one. No error is raised, but the previous data associated with that key is lost.**

5. In your own words, explain the concept of a key-value pair.

Answer: **A key-value pair is a custom data association where the "key" acts as a unique, descriptive label or index name, and the "value" is the actual data container attached to it. It is like looking up a specific word (the key) in a physical book to find its description (the value).**
