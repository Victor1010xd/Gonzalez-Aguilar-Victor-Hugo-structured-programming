A system that calculates the average grade of students and determines if they pass or fail.

# 1. 1. Imperative Programming  

## Brief Explanation
Modular programming takes structured programming a step further by breaking the program down into separate, independent modules (often different files or distinct logical blocks). Each module has a specific responsibility (e.g., data, logic, or user interface), making the code highly reusable, easier to maintain, and cleaner to scale.

## Code Example
Complete the steps:

````python
class GradesOperations:
    @staticmethod
    def calculate_average(grades_list):
        if not grades_list:
            return 0
        return sum(grades_list) / len(grades_list)

    @staticmethod
    def check_passing(average, passing_score=70):
        return "Pass" if average >= passing_score else "Fail"
def main():
    student_grades = [85, 92, 78, 60]
    avg = GradesOperations.calculate_average(student_grades)
    status = GradesOperations.check_passing(avg)

    print(f"Average: {avg:.2f}")
    print(f"Status: {status}")

if __name__ == "__main__":
    main()
````

## Your Modification
What did you change and why?

I added a secondary configuration or helper module block to dynamically adjust the passing grade threshold. I did this to demonstrate how modular code allows you to change business rules in one place without touching the core calculation logic.


## Reflection
What did you learn from this approach and how does it compare to others?
I learned that modular programming enforces boundaries. While structured programming groups code into functions, modular programming groups those functions into logical containers. Compared to imperative programming, it is much cleaner and prevents the main script from becoming cluttered as the project grows.

## 2. Structured Programming (functions + control structures)






## Brief Explanation
OOP mirrors the real world by organizing code around "objects" rather than actions. Objects combine data (attributes) and behaviors (methods) into a single blueprint called a Class. This allows us to treat a student and their grades as a self-contained entity.
## Code Example
Complete the steps:

````python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades
        self.average = 0
        self.status = "Pending"

    def process_academic_standing(self):
        if self.grades:
            self.average = sum(self.grades) / len(self.grades)
        self.status = "Pass" if self.average >= 70 else "Fail"

    def display_report(self):
        print(f"Student: {self.name}")
        print(f"Average: {self.average:.2f}")
        print(f"Status: {self.status}")
student1 = Student("Victor Hugo", [75, 88, 91])
student1.process_academic_standing()
student1.display_report()
````

## Your Modification
What did you change and why?
I added a new attribute self.name to the constructor and created a display_report() method inside the class. I did this to fully embrace the OOP paradigm, ensuring the object handles its own data presentation rather than relying on global print statements.


## Reflection
What did you learn from this approach and how does it compare to others?
OOP completely changes the mindset. Instead of thinking about variables passing through functions, I thought about an entity (the Student) that manages its own state. It feels much safer and more organized than imperative or structured programming because the data is protected inside the object.

## Brief Explanation
Functional programming treats computation as the evaluation of pure mathematical functions and avoids changing-state or mutable data. It heavily relies on built-in higher-order functions like map(), filter(), and reduce(), using concise expressions (like lambdas) to transform data cleanly.
## Code Example
Complete the steps:

````python
from functools import reduce
calculate_total = lambda list_data: reduce(lambda x, y: x + y, list_data)
calculate_avg   = lambda list_data: calculate_total(list_data) / len(list_data) if list_data else 0
determine_status = lambda average: "Pass" if average >= 70 else "Fail"
grades = [95, 82, 88]
avg = calculate_avg(grades)
status = determine_status(avg)
print("Average: ", avg)
print("Status: ", status)
````

## Your Modification
What did you change and why?
I utilized the reduce function from the functools library combined with a lambda expression to calculate the sum, instead of using Python's built-in sum() function. I did this to strictly follow functional programming principles by avoiding loop states and demonstrating explicit mathematical reduction.

## Reflection
What did you learn from this approach and how does it compare to others?
Functional programming is incredibly compact. Because it eliminates loops and variable mutations, the risk of side-effect bugs is dramatically reduced. Compared to the step-by-step nature of imperative programming or the heavy setup of OOP, functional programming feels like writing a series of mathematical equations that data just flows through.
