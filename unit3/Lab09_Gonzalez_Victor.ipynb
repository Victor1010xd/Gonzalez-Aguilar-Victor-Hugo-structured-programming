# 1. Introduction

A function that calls itself is called a **recursive function**.

Recursion is based on the following idea:

> A complex problem can sometimes be solved by reducing it to a smaller version of the same problem.

This process continues until a simple case is reached.

## The Two Essential Components of Recursion

### Base Case
The condition that stops the recursion.

### Recursive Case
The part of the function that calls itself using a smaller version of the problem.

### General Form
```python
def function(problem):
    if simple_case:
        # base case
        do_something
    else:
        # recursive case
        function(smaller_problem)
```

# 2. First Recursive Program

````python
def countdown(n):
    if n == 0:
        print("Liftoff!")
    else:
        print(n)
        countdown(n - 1)

countdown(5)
````

### Questions
1. Which statement makes this function recursive?
The line countdown(n - 1)
---

2. What is the base case?
The condition if n == 0 It prints "Liftoff!" and stops execution.
---

3. Why does the function eventually stop?
The variable n decreases each time. It subtracts one during every call. It reaches zero and hits the base case.
---

4. What value is passed during each new function call?
The value of n - 1. This represents the current number minus one.
---

# 3. Understanding Function Calls

When a recursive function calls itself, Python temporarily pauses the current function call and starts executing the new one.

After the base case is reached, the function calls finish one by one in reverse order. This behavior is known as the **call stack**.

````python
def explore(level):
    print("Entering level", level)

    if level > 1:
        explore(level - 1)

    print("Leaving level", level)

explore(3)
````

Predict the output.

Write the actual output.

### Questions
1. Which message appears first?
"Entering level 3". This is printed when explore(3) is initially called.

2. Which message appears last?
"Leaving level 3". This is printed right before the initial function call completely finishes.

3. Why are the 'Leaving' messages printed in reverse order?
They use the call stack mechanism. The most recent function call must finish first. Paused calls resume only after newer calls exit.

# 5. Returning Values Recursively

````python
def treasures(level):
    if level == 0:
        return 0
    else:
        return 1 + treasures(level - 1)

print(treasures(4))
````

Trace the execution:

```text
treasures(4)
= 1 + treasures(3)
= 1 + 1 + treasures(2)
= 1 + 1 + 1 + treasures(1)
= 1 + 1 + 1 + 1 + treasures(0)
= 1 + 1 + 1 + 1 + 0
= 4

```

# 6. Recursive Summation

````python
def coins(level):
    if level == 0:
        return 0
    else:
        return level + coins(level - 1)

print(coins(5))
````

Determine the value returned by `coins(5)` and show every step of the recursive process.

coins(5)
= 5 + coins(4)
= 5 + 4 + coins(3)
= 5 + 4 + 3 + coins(2)
= 5 + 4 + 3 + 2 + coins(1)
= 5 + 4 + 3 + 2 + 1 + coins(0)
= 5 + 4 + 3 + 2 + 1 + 0
= 15

# 8. Common Errors in Recursive Programming

````python
def mystery(n):
    if n == 0:
        return
    print(n)
    mystery(n - 1)

mystery(5)
````

### Questions
1. Does the function stop?

No. It runs indefinitely.

2. What error is eventually generated?

RecursionError. It states maximum recursion depth exceeded.

3. Why does this error occur?

There is no base case. The call stack runs out of memory.

4. How can the function be corrected?

Add a condition to stop the recursion, for example: if n == 0: return

# 9. Recursion versus Iteration

| Iteration | Recursion |
|-----------|-----------|
| Uses loops. | Uses function calls. |
| Usually requires less memory. | Often produces simpler solutions. |
| Execution is generally faster. | Can model naturally recursive problems. |

Examples of naturally recursive problems include:
- traversing folders inside folders;
- exploring menus and submenus;
- searching family trees;
- processing nested data structures.
